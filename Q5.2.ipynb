```python
import numpy as np
from gensim.models import KeyedVectors
import gensim.downloader as api
```

```python
print("Hello")
```

```python
i = 10
j = 0
```

```python
model = api.load("word2vec-google-news-300")
```

```python
def cosine_similarity(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
```

```python
def check(words: list):
    w1, w2, w3, w4 = words

    # Skip if any word not in vocabulary
    if any(word not in model for word in [w1, w2, w3, w4]):
        print("Skipped: not in vocabulary")
        return

    vec = model[w2] - model[w1] + model[w3]

    similarity = cosine_similarity(vec, model[w4])
    
    print(f"Similarity between predicted and actual word is: {similarity}\n Word Found: {model.most_similar(positive=[vec],topn=1)[0][0]}\n")
```

```python
Input_File = open("questions-words.txt", "r", encoding="utf-8")
while j != i:
    sentence = Input_File.readline()

    if sentence == "":
        break

    if sentence.startswith(":"):
        continue

    print(sentence.strip())

    words = sentence.strip().split()

    if len(words) == 4:
        check(words)
        j += 1
Input_File.close()
```
